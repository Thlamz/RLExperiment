```python
import pandas as pd
import seaborn as sns
```

```python
progress_no_intention = pd.read_csv("./progress.csv")
progress_no_intention["intention"] = 0
progress_1_intention = pd.read_csv("./progress_1.csv")
progress_1_intention["intention"] = 1

data=pd.concat([progress_no_intention, progress_1_intention])

print("\n".join(data.columns))
```

```python
ax = sns.lineplot(data, x="time_total_s", y="custom_metrics/win_mean", hue="intention")

ax.set(ylabel="Win rate")
ax.set(xlabel="Training Time")
# Make graph wider
import matplotlib.pyplot as plt
plt.gcf().set_size_inches(15, 8)
```
